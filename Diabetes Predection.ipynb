Importing Depedences

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Data Collection And Analysis

```python
data = pd.read_csv('/content/diabetes.csv')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
# menampilkan data baris dan kolom
data.shape
```

```python
data ['Outcome'].value_counts()
```

0 --> Non-Diabetic

1 --> Diabetic

```python
data.groupby('Outcome').mean()
```

```python
 #separating the data and labels
X = data.drop(columns = 'Outcome',axis = 1)
Y = data['Outcome']
```

```python
print(X)
```

```python
print(Y)
```

Data Standarlization

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
standardized_data = scaler.transform(X)
```

```python
print(standardized_data)
```

```python
X = standardized_data
Y = data['Outcome']
```

```python
print(X)
print(Y)
```

Trainng Test Split


```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,stratify =Y, random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Training The Model

```python
classifier = svm.SVC(kernel = 'linear')
```

```python
#Training the support vector Machine Classifier
classifier.fit(X_train,Y_train)
```

Model Evaluasi

Accuracy Score

```python
# accuracy score on the training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy Score Of The Training Data : ',training_data_accuracy)
```

```python
# accuracy score on the test data
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print("Accuracy Score of the data training :",test_data_accuracy)
```

Making A Predictive System

```python
input_data = (0,137,40,35,168,43.1,2.288,33)

# change the input_data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the array as we are predicting for one instance
input_data_reshape = input_data_as_numpy_array.reshape(1,-1)

# standarlize the input data
std_data = scaler.transform(input_data_reshape)
print(std_data)

prediction = classifier.predict(std_data)
print(prediction)

if(prediction[0]== 0):
  print('The Prediction No diabetic')
else:
  print('The Person is a Diabetic')

```
