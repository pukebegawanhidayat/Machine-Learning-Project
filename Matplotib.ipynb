```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

Ploting Data

```python
# sine wave
plt.plot(x,y)
plt.show()
```

```python
# cosine wave
plt.plot(x,z)
plt.show()
```

```python
# adding title x-axis and y-axis labels
plt.plot(x,y)
plt.xlabel('Angle')
plt.ylabel('sine Value')
plt.title('sine Wave')
plt.show()
```

```python
# parabola
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y ,'r+')
plt.show()
```

Bar Plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English','Frances','Spanish','Latin','Germany']
people = [100,50,150,40,70]
ax.bar(languages,people)
plt.xlabel('Languages')
plt.ylabel('Number Of People')
plt.show
```

Pie Chart

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English','Frances','Spanish','Latin','Germany']
people = [100,50,150,40,70]
ax.pie(people,labels=languages,autopct='%1.1f%%')
plt.show
```

scatter plot


```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show
```

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
```
